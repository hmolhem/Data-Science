## Arithmetic methods with fill values

```python
arr1 = np.arange(12).reshape((4,3))
df1 = pd.DataFrame(data=arr1, columns=list('abc'))
df1
```

```python
arr2 = np.arange(10).reshape((5,2))
df2 = pd.DataFrame(data=arr2, columns=list('ab'))
df2
```

```python
df1.loc[1,'b'] = np.nan
df1
```

```python
# 2+NaN = NaN
df1+df2
```

```python
df1.add(df2)
```

```python
# NaN can be o by this switch
df1.add(df2, fill_value=0)
```

```python
df1.sub(df2)
```

```python
arr1 = np.arange(12).reshape((4,3))
df1 = pd.DataFrame(data=arr1, columns=list('abc'))
df1
```

```python
df1.cumsum()
```

## Hierarchical indexing

```python
frame = pd.DataFrame({
    'C++': [6, 7, 4, 16, 14, 13, 15, 7],
    'Python': [3, 2, 5, 14, 19, 18, 16, 10],
    'K1': ['one', 'one', 'one', 'two', 'two','two','two','two'],
    'K2': ['ali', 'reza', 'sara', 'ali', 'reza', 'sara', 'taha', 'farid']
})
frame
```

```python
df = frame.set_index(['K1', 'K2'])
df
```

```python
df.sort_index(level=1)
```

```python
df.mean(level='K2')
```

```python
df
```

```python
df.unstack()
```

```python
df.swaplevel('K1','K2')
```
